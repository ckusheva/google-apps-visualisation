```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
df1 = pd.read_csv('googleplaystore.csv')
plt.plot(df1['Rating']) 
plt.subplot(1, 2, 2)
df1 = df1.drop(df1.index[df1.Rating >5])
plt.plot(df1['Rating'])
plt.show
```

```python
df1.columns
Category_Rating_Mean = df1.groupby('Category').aggregate(np.mean)
print(Category_Rating_Mean)
```

```python
Top_10 = Category_Rating_Mean.sort_values('Rating', ascending=False).head(10)
print(Top_10)
```

```python
Top_10['Rating']
```

```python
Top_10.plot(kind = 'bar', title = 'Top 10 Mean Ratings by Catgory')
plt.ylabel('Rating')
plt.show()
```
